```python
import requests  # Библиотека для запроса

a = 1  # Переменная для перехода по страницам
while a <= 14:
    # Объявление переменных как глобальные
    global bodyType, brand, color, complectation_dict, engineDisplacement
    global enginePower, equipment_dict, fuelType, image, mileage, modelDate
    global model_name, name, numberOfDoors, priceCurrency, productionDate
    global sell_id, super_gen, vehicleTransmission, vendor, owners_number
    global pts, gear_type, steering_wheel, custom_cleared, price

    # URL на который будет отправлен запрос
    URL = 'https://auto.ru/-/ajax/desktop/listing/'

    params = {
        "category": "cars",
        "section": "used",
        "price_to": 6000000,
        "catalog_filter": [{
            "mark": "LEXUS"
        }],
        "page": a,
        "geo_id": [213]
    }
    #Заголовки страницы
    HEADERS = {
        'Accept':
        '*/*',
        'Accept-Encoding':
        'gzip, deflate, br',
        'Accept-Language':
        'ru-RU,ru;q=0.8,en-US;q=0.5,en;q=0.3',
        'Connection':
        'keep-alive',
        'Content-Length':
        '115',
        'Cookie':
        'autoru_sid=a%3Ag60221bed21vslp9agvo99l4t484kvda.20cf7ac71bfa4b13e6c76d07865ded90%7C1612848109192.604800.KjOclrhyyMgaGS60pWmcoA.kQRO4Y-qnsIA8u-g-gUGlZxdYw3vXjlFZXa5KkvicdM; '
        'autoruuid=g60221bed21vslp9agvo99l4t484kvda.20cf7ac71bfa4b13e6c76d07865ded90; '
        'suid=047fd897015e89b934392c2c0d6c2e43.5be655abcb81e1b6ea386c9090aa6fde; '
        'yuidlt=1; yandexuid=9312377611612848109; counter_ga_all7=2; '
        'crookie=QlCoWyX4wtZfmnHnIDVB3NEvSk43nWnQMIfu/yvlvE7YrUuGlNXZOmQlp7EunVYwASN4Ck3A3HZ3z+cw/Tfl5C+Y2sU=; '
        'cmtchd=MTYxMjg0ODEwODQ2OA==; _ga=GA1.2.554774267.1612848109; '
        '_gid=GA1.2.1072661525.1612848109; _ym_uid=1612848109671562223; '
        '_ym_d=1612984416; gids=213; '
        'cycada=fu+oxiFm4ksqVtippRKRK64lZWk8I9ZZr5MVTzO1zyo=; '
        '_csrf_token=663e57939f148764fb84ad0f97c70b4bf853ca3250d296e4; '
        'gdpr=0; from=direct; from_lifetime=1612984416389; '
        'autoru-visits-count=2; gradius=0; _ym_isad=2; X-Vertis-DC=vla',
        'Host':
        'auto.ru',
        'Origin':
        'https://auto.ru',
        'Referer':
        'https://auto.ru/moskva/cars/lexus/used/?price_to=6000000&page={a}',
        'User-Agent':
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:85.0) '
        'Gecko/20100101 Firefox/85.0',
        'content-type':
        'application/json',
        'x-client-app-version':
        'd954678372',
        'x-client-date':
        '1612984426722',
        'x-csrf-token':
        '663e57939f148764fb84ad0f97c70b4bf853ca3250d296e4',
        'x-page-request-id':
        '1f19ada660890819945086627714415d',
        'x-requested-with':
        'fetch'
    }

    response = requests.post(URL, json=params, headers=HEADERS)
    # Делаем post запрос на url
    data = response.json()['offers']
    # Переменная data хранит полученные объявления

    i = 0  # Переменная для перехода по объявлениям
    while i <= len(
            data) - 1:  # len(data)-1 это количество пришедших объявлений

        # bodyType
        try:
            bodyType = str(
                data[i]['vehicle_info']['configuration']['human_name'])
        except:
            bodyType = 'NaN'

        # brand - [vehicle_info][mark_info][code]
        try:
            brand = str(data[i]['vehicle_info']['mark_info']['code'])
        except:
            brand = 'NaN'

        # Цвет автомобиля (возвращается в формате hex)
        try:
            color = str(data[i]['color_hex'])
        except:
            color = 'Not color'

    # complectation_dict - [complectation]
        try:
            complectation_dict = str(data[i]['vehicle_info']['complectation'])
        except:
            complectation_dict = 'NaN'

        # engineDisplacement
        try:
            engineDisplacement = str(
                data[i]['vehicle_info']['tech_param']['human_name'])
        except:
            engineDisplacement = 'NaN'

        # eenginePower
        try:
            enginePower = str(data[i]['vehicle_info']['tech_param']['power'])
        except:
            enginePower = 'NaN'

        # equipment_dict
        try:
            equipment_dict = str(data[i]['vehicle_info']['equipment'])
        except:
            equipment_dict = 'NaN'

        # fuelType
        try:
            fuelType = str(
                data[i]['vehicle_info']['tech_param']['engine_type'])
        except:
            fuelType = 'NaN'

        # image
        try:
            image = str(data[i]['vehicle_info']['configuration']['main_photo']
                        ['sizes']['orig'])
        except:
            image = 'NaN'

        # Пробег автомобиля
        try:
            mileage = str(data[i]['state']['mileage'])
        except:
            mileage = 'NaN'

        # modelDate
        try:
            modelDate = str(data[i]['vehicle_info']['super_gen']['year_from'])
        except:
            modelDate = 'NaN'

        # model_name
        try:
            model_name = str(data[i]['vehicle_info']['model_info']['code'])
        except:
            model_name = 'NaN'

        # name
        try:
            name = str(data[i]['vehicle_info']['tech_param']['human_name'])
        except:
            name = 'NaN'

        # numberOfDoors
        try:
            numberOfDoors = str(
                data[i]['vehicle_info']['configuration']['doors_count'])
        except:
            numberOfDoors = 'NaN'

        # priceCurrency
        try:
            priceCurrency = str(data[i]['price_info']['currency'])
        except:
            priceCurrency = 'NaN'

        # productionDate
        try:
            productionDate = str(data[i]['documents']['year'])
        except:
            productionDate = 'NaN'

        # sell_id
        try:
            sell_id = str(data[i]['id'])
        except:
            sell_id = 'NaN'

        # super_gen
        try:
            super_gen = str(data[i]['vehicle_info']['super_gen'])
        except:
            super_gen = 'NaN'

        # vehicleTransmission
        try:
            vehicleTransmission = str(
                data[i]['vehicle_info']['tech_param']['transmission'])
        except:
            vehicleTransmission = 'NaN'

        # vendor
        try:
            vendor = str(data[i]['vehicle_info']['vendor'])
        except:
            vendor = 'NaN'

        # Владельцы
        try:
            owners_number = str(data[i]['documents']['owners_number'])
        except:
            owners_number = 'NaN'

        # PTS автомобиля
        try:
            pts = str(data[i]['documents']['pts_original'])
        except:
            pts = 'False'

        # Привод gear_type
        try:
            gear_type = str(data[i]['vehicle_info']['tech_param']['gear_type'])
        except:
            gear_type = 'NaN'

        # Руль steering_wheel
        try:
            steering_wheel = str(data[i]['vehicle_info']['steering_wheel'])
        except:
            steering_wheel = 'NaN'

        try:
            custom_cleared = str(data[i]['documents']['custom_cleared'])
        except:
            custom_cleared = 'NaN'

        # Цена
        try:
            price = str(data[i]['price_info']['price'])
        except:
            price = 'NaN'

        # Переменная в которую всё записываем
        text = str(bodyType) + ';' + str(brand) + ';' + str(color) + ';' + str(complectation_dict) + ';' + \
            engineDisplacement + ';' + enginePower + ';' + equipment_dict + ';' + fuelType + ';' + \
            image + ';' + mileage + ';' + modelDate + ';' + model_name + ';' + name + ';' + numberOfDoors + ';' + \
            priceCurrency + ';' + productionDate + ';' + sell_id + ';' + super_gen + ';' + \
            vehicleTransmission + ';' + vendor + ';' + owners_number + ';' + pts + ';' + gear_type + ';' + \
            str(steering_wheel) + ';' + str(custom_cleared) + ';' + str(price) + '\n'

        # Записываем переменную в файл
        with open('Save_auto1.txt', 'a', encoding='UTF-8') as file:
            # Открываем файл Save_auto1.txt (создаётся автоматически), на дозапись (ключ a) encoding='UTF-8'
            file.write(text)  # Записываем переменную
        i += 1  # Увеличиваем переменную перехода по объявлениям на 1
    print('Page: ' + str(a))  # Выводим сообщение, какая страница записалась
    a += 1  # Увеличиваем переменную страницы сайта на 1

print('Successfully')  # Выводим информацию об успешном выполнении
```

Итоговые результаты записал в файл Save_auto1.txt через ";". Разме файла 42 Mb, что позволило комфортно отработать на персональном компьютере.
